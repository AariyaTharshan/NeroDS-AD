## Define augmentation function

### Subtask:
Create a function to apply a specific number of consistent augmentations to a pair of images and save each augmented pair as a new image.


**Reasoning**:
Define a function to apply consistent augmentation to paired images and save them.



```python
from torchvision import transforms
import random

def apply_consistent_augmentation(mri_path, pet_path, mri_output_dir, pet_output_dir, num_augmentations=9):
    """
    Applies consistent augmentation to a pair of MRI and PET images and saves them.

    Args:
        mri_path (str): Path to the input MRI image.
        pet_path (str): Path to the input PET image.
        mri_output_dir (str): Directory to save augmented MRI images.
        pet_output_dir (str): Directory to save augmented PET images.
        num_augmentations (int): Number of augmentations to apply.
    """
    try:
        mri_img = Image.open(mri_path)
        pet_img = Image.open(pet_path)

        # Define augmentation transformations
        augmentations = transforms.Compose([
            transforms.RandomRotation(degrees=30),   # stronger rotation
            transforms.RandomHorizontalFlip(p=0.5), # Increased probability
            transforms.RandomVerticalFlip(p=0.5),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1), # Increased jitter
            transforms.RandomAffine(degrees=0, scale=(0.8, 1.2), shear=(-10, 10, -10, 10)), # Added affine transformation
            transforms.RandomResizedCrop(size=desired_size, scale=(0.6, 1.0), ratio=(0.75, 1.25)), # Adjusted scale and ratio
            transforms.GaussianBlur(kernel_size=(3,3), sigma=(0.1, 1.0))
        ])


        for i in range(num_augmentations):
            # Set a random seed to ensure consistent augmentation
            seed = random.randint(0, 100000)
            random.seed(seed)
            torch.manual_seed(seed)

            augmented_mri = augmentations(mri_img)

            random.seed(seed)
            torch.manual_seed(seed)

            augmented_pet = augmentations(pet_img)

            # Save augmented images
            base_filename = os.path.splitext(os.path.basename(mri_path))[0]
            mri_output_path = os.path.join(mri_output_dir, f"{base_filename}_aug_{i+1}.png")
            pet_output_path = os.path.join(pet_output_dir, f"{base_filename}_aug_{i+1}.png")

            augmented_mri.save(mri_output_path)
            augmented_pet.save(pet_output_path)

            print(f"Saved augmented pair {i+1} for {base_filename}")

        mri_img.close()
        pet_img.close()

    except FileNotFoundError:
        print(f"Error: One or both image files not found: {mri_path}, {pet_path}")
    except Exception as e:
        print(f"Error processing image pair {os.path.basename(mri_path)} and {os.path.basename(pet_path)}: {e}")
```

### Subtask:

**Reasoning**:
Iterate through the filenames in the MRI input directory, construct the corresponding PET path, and call the augmentation function for each pair.



```python
import os
import time
import torch
import matplotlib.pyplot as plt

num_augmentations = 9

# Metrics storage
mri_success = 0
pet_success = 0
missing_pet = 0
processing_times = []
augment_counts = []

for filename in os.listdir(normalized_mri_input_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        mri_input_image_path = os.path.join(normalized_mri_input_path, filename)
        pet_input_image_path = os.path.join(normalized_pet_input_path, filename)  # Matching MRI-PET assumption

        if os.path.exists(pet_input_image_path):
            try:
                start_time = time.time()

                apply_consistent_augmentation(
                    mri_input_image_path,
                    pet_input_image_path,
                    augment_mri_output_path,
                    augment_pet_output_path,
                    num_augmentations
                )

                end_time = time.time()
                processing_times.append(end_time - start_time)

                # Track metrics
                mri_success += 1
                pet_success += 1
                augment_counts.append(num_augmentations)

                print(f"Augmented MRI & PET for: {filename}")

            except Exception as e:
                print(f"Error augmenting {filename}: {e}")

        else:
            missing_pet += 1
            print(f"Warning: Corresponding PET file not found for {filename}")

# ----- Metrics Summary -----
print("\n===== Augmentation Summary =====")
print(f"Total MRI files scanned: {mri_success + missing_pet}")
print(f"Augmented MRI-PET pairs: {mri_success}")
print(f"Missing PET files: {missing_pet}")
if processing_times:
    print(f"Average Processing Time per Pair: {sum(processing_times)/len(processing_times):.4f} seconds")
    print(f"Total Augmentations Generated: {sum(augment_counts)}")

# ----- Visualization -----
plt.figure(figsize=(12, 5))

# 1. Success vs Missing PET
plt.subplot(1, 2, 1)
plt.bar(["Augmented Pairs", "Missing PET"], [mri_success, missing_pet], color=["green", "red"])
plt.title("MRI-PET Pair Processing Outcome")
plt.ylabel("Count")

# 2. Histogram of augmentation counts
plt.subplot(1, 2, 2)
plt.hist(augment_counts, bins=10, color="blue", alpha=0.7)
plt.title("Augmentation Count per File")
plt.xlabel("Number of Augmentations")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Define the cnn + transformer model architecture

### Subtask:
Create a model that combines a Convolutional Neural Network (CNN) for initial feature extraction and a Transformer model for processing sequential features.


**Reasoning**:
Define a combined CNN and Transformer model for feature extraction.



```python
import torch
import torch.nn as nn
import torchvision.models as models

class CNNTransformerFeatureExtractor(nn.Module):
    def __init__(self, num_channels=3, cnn_out_channels=32,
                 num_heads=8, num_layers=2, pretrained=True):
        super(CNNTransformerFeatureExtractor, self).__init__()

        # 1️⃣ Custom CNN branch (lightweight, task-specific)
        self.custom_cnn = nn.Sequential(
            nn.Conv2d(num_channels, cnn_out_channels, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(cnn_out_channels),
            nn.ReLU(inplace=True),

            nn.Conv2d(cnn_out_channels, cnn_out_channels * 2, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(cnn_out_channels * 2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),

            nn.Conv2d(cnn_out_channels * 2, cnn_out_channels * 4, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(cnn_out_channels * 4),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((7, 7))
        )
        self.custom_out_channels = cnn_out_channels * 4

        # 2️⃣ Pretrained EfficientNet-B0 branch
        backbone = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1 if pretrained else None)
        self.pretrained_cnn = nn.Sequential(*list(backbone.features), nn.AdaptiveAvgPool2d((7, 7)))
        self.pretrained_out_channels = 1280

        # 3️⃣ Combined output channels
        self.total_channels = self.custom_out_channels + self.pretrained_out_channels

        # 4️⃣ Positional embedding for Transformer
        self.pos_embedding = nn.Parameter(torch.randn(1, 49, self.total_channels))

        # 5️⃣ Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=self.total_channels,
            nhead=num_heads,
            batch_first=True,
            dropout=0.1,
            dim_feedforward=2048
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

    def forward(self, x):
        # Custom CNN features
        custom_feat = self.custom_cnn(x)  # (B, C1, 7, 7)
        # Pretrained CNN features
        pretrained_feat = self.pretrained_cnn(x)  # (B, C2, 7, 7)

        # Concatenate along channel dimension
        combined_feat = torch.cat([custom_feat, pretrained_feat], dim=1)  # (B, C1+C2, 7, 7)
        B, C, H, W = combined_feat.shape

        # Flatten spatial dimensions for Transformer
        transformer_input = combined_feat.view(B, C, H * W).permute(0, 2, 1)  # (B, 49, C)

        # Add positional embedding
        transformer_input = transformer_input + self.pos_embedding[:, :transformer_input.size(1)]

        # Transformer Encoder
        transformer_output = self.transformer_encoder(transformer_input)

        # Mean pooling over tokens
        pooled_features = transformer_output.mean(dim=1)  # (B, total_channels)

        return pooled_features
```

## Define Classification Head

### Subtask:
Create a classification head using fully connected layers that takes the combined features as input.

**Reasoning**:
Define a simple classification head with fully connected layers that takes the combined features as input and outputs a prediction.

```python
import torch.nn as nn

class ClassificationHead(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(ClassificationHead, self).__init__()
        self.fc1 = nn.Linear(input_dim, 256)
        self.bn1 = nn.BatchNorm1d(256)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(p=0.3)
        
        self.fc2 = nn.Linear(256, 128)
        self.bn2 = nn.BatchNorm1d(128)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(p=0.2)
        
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        
        x = self.fc2(x)
        x = self.bn2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        
        x = self.fc3(x)
        return x

# Assuming combined_features_array is the numpy array of your combined features
# input_dim will be the number of features in combined_features_array
input_dim = combined_features_array.shape[1]
num_classes = 3

classification_head = ClassificationHead(input_dim, num_classes)
classification_head.to(device)
print(classification_head)
```

### Subtask:
Convert the combined features and corresponding labels into PyTorch tensors and create data loaders.

Convert the numpy arrays of combined features and labels into PyTorch tensors and create data loaders for efficient batch processing during training.

```python
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
import torch
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Generate pseudo-labels from clustering (NO manual labels needed!)
num_samples = combined_features_array.shape[0]

if num_samples > 0:
    # Standardize features for better clustering
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(combined_features_array)
    
    # Find optimal number of clusters
    best_score = -1
    best_n_clusters = 3
    for n in range(2, 6):
        kmeans = KMeans(n_clusters=n, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(features_scaled)
        score = silhouette_score(features_scaled, cluster_labels)
        if score > best_score:
            best_score = score
            best_n_clusters = n
    
    # Generate pseudo-labels
    kmeans = KMeans(n_clusters=best_n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(features_scaled)
    
    print(f"Generated {best_n_clusters} clusters (silhouette score: {best_score:.4f})")
    unique, counts = np.unique(labels, return_counts=True)
    for label, count in zip(unique, counts):
        print(f"  Cluster {label}: {count} samples ({count/len(labels)*100:.1f}%)")

    # Convert to PyTorch tensors
    combined_features_tensor = torch.from_numpy(combined_features_array).float()
    labels_tensor = torch.from_numpy(labels).long()

    # Create a custom Dataset
    class PETMRIDataset(Dataset):
        def __init__(self, features, labels):
            self.features = features
            self.labels = labels

        def __len__(self):
            return len(self.features)

        def __getitem__(self, idx):
            return self.features[idx], self.labels[idx]

    # Create dataset and split into train/val/test
    dataset = PETMRIDataset(combined_features_tensor, labels_tensor)
    
    train_size = int(0.70 * len(dataset))
    val_size = int(0.15 * len(dataset))
    test_size = len(dataset) - train_size - val_size
    
    train_dataset, val_dataset, test_dataset = random_split(
        dataset, [train_size, val_size, test_size],
        generator=torch.Generator().manual_seed(42)
    )
    
    # Create DataLoaders
    batch_size = 32
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    print(f"\nTrain: {len(train_dataset)} samples ({len(train_loader)} batches)")
    print(f"Val:   {len(val_dataset)} samples ({len(val_loader)} batches)")
    print(f"Test:  {len(test_dataset)} samples ({len(test_loader)} batches)")
else:
    print("No combined features available. Please ensure feature extraction and combination were successful.")
```

## Apply Grad-CAM for Explainability

### Subtask:
Implement Grad-CAM to visualize the important regions in the input images that contribute to the model's predictions.

**Reasoning**:
Implement Grad-CAM to generate heatmaps highlighting the salient regions in the input images based on the gradients flowing through the final convolutional layer of the trained model.

```python
import torch.nn.functional as F
import matplotlib.pyplot as plt
import torch
import cv2
import numpy as np

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None

        # Register hooks to capture gradients and activations
        self.target_layer.register_forward_hook(self.save_activation)
        self.target_layer.register_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activations = output

    def save_gradient(self, module, grad_in, grad_out):
        self.gradients = grad_out[0]

    def __call__(self, input_tensor, target_category=None):
        self.model.zero_grad()
        output = self.model(input_tensor)

        if target_category is None:
            target_category = torch.argmax(output, dim=1)

        one_hot = torch.zeros_like(output)
        one_hot.scatter_(1, target_category.unsqueeze(1), 1.)

        output.backward(gradient=one_hot, retain_graph=True)

        guided_gradients = self.gradients

        # Calculate weights (alpha) for each feature map
        weights = torch.mean(guided_gradients, dim=(2, 3), keepdim=True)

        # Calculate Grad-CAM heatmap
        heatmap = torch.sum(weights * self.activations, dim=1, keepdim=True)

        # Apply ReLU to the heatmap
        heatmap = F.relu(heatmap)

        # Upsample heatmap to the size of the input image
        heatmap = F.interpolate(heatmap, size=(input_tensor.size(2), input_tensor.size(3)), mode='bilinear', align_corners=False)

        # Normalize heatmap
        heatmap = heatmap / (torch.max(heatmap) + 1e-8)  # Add epsilon to avoid division by zero

        return heatmap

def show_cam_on_image(img, heatmap):
    # --- Convert image to HWC NumPy ---
    if isinstance(img, torch.Tensor):
        if img.ndim == 4:
            img = img[0]
        img = img.permute(1, 2, 0).cpu().detach().numpy()
    
    # Normalize image to [0,1]
    img = img - np.min(img)
    img = img / (np.max(img) + 1e-8)

    # --- Convert heatmap to NumPy HxW ---
    if isinstance(heatmap, torch.Tensor):
        if heatmap.ndim == 4:
            heatmap = heatmap[0, 0]
        elif heatmap.ndim == 3:
            heatmap = heatmap[0]
        heatmap = heatmap.cpu().detach().numpy()

    # Resize heatmap to match image
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Overlay heatmap
    cam = heatmap * 0.4 + np.uint8(255 * img)
    cam = np.uint8(cam)
    return cam

# Select a sample image pair
batch_index = 0
image_index = 0

# Ensure batches exist
if ('loaded_mri_batches' in locals() and 'loaded_pet_batches' in locals() and
    batch_index < len(loaded_mri_batches) and
    image_index < loaded_mri_batches[batch_index].shape[0]):

    sample_mri = loaded_mri_batches[batch_index][image_index]
    sample_pet = loaded_pet_batches[batch_index][image_index]

    # Convert to tensor and move to device
    sample_mri_tensor = torch.from_numpy(sample_mri).permute(2, 0, 1).float().unsqueeze(0).to(device)
    sample_pet_tensor = torch.from_numpy(sample_pet).permute(2, 0, 1).float().unsqueeze(0).to(device)

    # --- Grad-CAM setup ---
    target_layer_custom = feature_extractor.custom_cnn[-3]
    target_layer_pretrained = feature_extractor.pretrained_cnn[-2]

    # Create GradCAM instances
    grad_cam_mri = GradCAM(feature_extractor, target_layer_custom)
    grad_cam_pet = GradCAM(feature_extractor, target_layer_custom)

    # Generate heatmaps
    mri_heatmap = grad_cam_mri(sample_mri_tensor)
    pet_heatmap = grad_cam_pet(sample_pet_tensor)

    # Convert tensors for overlay
    mri_img_np = sample_mri_tensor[0].permute(1, 2, 0).cpu().numpy()
    pet_img_np = sample_pet_tensor[0].permute(1, 2, 0).cpu().numpy()

    # Generate Grad-CAM overlays
    mri_cam_image = show_cam_on_image(mri_img_np, mri_heatmap)
    pet_cam_image = show_cam_on_image(pet_img_np, pet_heatmap)

    # Display results
    fig, axes = plt.subplots(2, 2, figsize=(12, 12))

    axes[0, 0].imshow(sample_mri)
    axes[0, 0].set_title("Original MRI", fontsize=14)
    axes[0, 0].axis('off')

    axes[0, 1].imshow(mri_cam_image)
    axes[0, 1].set_title("MRI Grad-CAM", fontsize=14)
    axes[0, 1].axis('off')

    axes[1, 0].imshow(sample_pet)
    axes[1, 0].set_title("Original PET", fontsize=14)
    axes[1, 0].axis('off')

    axes[1, 1].imshow(pet_cam_image)
    axes[1, 1].set_title("PET Grad-CAM", fontsize=14)
    axes[1, 1].axis('off')

    plt.tight_layout()
    plt.savefig('gradcam_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("✓ Grad-CAM visualization saved as 'gradcam_visualization.png'")

else:
    print("Could not load sample data. Check your batched MRI/PET datasets.")
```
